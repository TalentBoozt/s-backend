```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("E:/PROJECTS/spring boot/Talentboozt-projects/s-backend/src/main/resources/training/events_data.csv")
dfc = pd.read_csv("E:/PROJECTS/spring boot/Talentboozt-projects/s-backend/src/main/resources/training/events_data_cleaned.csv", dtype={'9': str})
```

```python
# original data frame
plt.figure(figsize=(12, 8))

# Plot distribution for selected columns
sns.histplot(df['durationMs'], kde=True, bins=50, color='skyblue', label='Duration (ms)')
sns.histplot(df['scrollPercent'], kde=True, bins=50, color='orange', label='Scroll Percent')

# Customize the plot
plt.title('Distribution of Duration and Scroll Percent')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
# cleaned data frame
plt.figure(figsize=(12, 8))

# Plot distribution for selected columns
sns.histplot(dfc['durationMs'], kde=True, bins=50, color='skyblue', label='Duration (ms)')
sns.histplot(dfc['scrollPercent'], kde=True, bins=50, color='orange', label='Scroll Percent')

# Customize the plot
plt.title('Distribution of Duration and Scroll Percent')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
# Plot countplot for 'eventType' (original data frame)
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='eventType', palette='viridis')

plt.title('Event Type Distribution')
plt.xlabel('Event Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plot boxplots for 'durationMs' and 'ttfb' to check for outliers (original data frame)
plt.figure(figsize=(12, 6))

# Boxplot for durationMs
plt.subplot(1, 2, 1)
sns.boxplot(x=df['durationMs'], color='skyblue')
plt.title('Boxplot of Duration (ms)')

# Boxplot for ttfb
plt.subplot(1, 2, 2)
sns.boxplot(x=df['ttfb'], color='orange')
plt.title('Boxplot of TTFB')

plt.tight_layout()
plt.show()
```

```python
# Plot boxplots for 'durationMs' and 'ttfb' to check for outliers (cleaned data frame)
plt.figure(figsize=(12, 6))

# Boxplot for durationMs
plt.subplot(1, 2, 1)
sns.boxplot(x=dfc['durationMs'], color='skyblue')
plt.title('Boxplot of Duration (ms)')

# Boxplot for ttfb
plt.subplot(1, 2, 2)
sns.boxplot(x=dfc['ttfb'], color='orange')
plt.title('Boxplot of TTFB')

plt.tight_layout()
plt.show()
```

```python
# Scatter plot for durationMs vs. scrollPercent (original data frame)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='durationMs', y='scrollPercent', color='purple')

plt.title('Scatter Plot: Duration vs. Scroll Percent')
plt.xlabel('Duration (ms)')
plt.ylabel('Scroll Percent')
plt.show()

# Pairplot for multiple numeric features
sns.pairplot(df[['durationMs', 'scrollPercent', 'downlink', 'rtt']])
plt.show()
```

```python
# Scatter plot for durationMs vs. scrollPercent (cleaned data frame)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=dfc, x='durationMs', y='scrollPercent', color='purple')

plt.title('Scatter Plot: Duration vs. Scroll Percent')
plt.xlabel('Duration (ms)')
plt.ylabel('Scroll Percent')
plt.show()

# Pairplot for multiple numeric features
sns.pairplot(dfc[['durationMs', 'scrollPercent', 'downlink', 'rtt']])
plt.show()
```

```python
# Pie chart for event type distribution (original data frame)
event_type_counts = df['eventType'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(event_type_counts, labels=event_type_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('Set3', n_colors=len(event_type_counts)))
plt.title('Event Type Distribution')
plt.show()
```

```python
# Correlation Heatmap of Numeric Features(original data frame)
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Compute the correlation matrix
corr = numeric_df.corr()

# Set up the plot
plt.figure(figsize=(12, 8))

# Create a heatmap of the correlation matrix
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)

# Customize the plot
plt.title('Correlation Heatmap of Numeric Features')
plt.show()
```

```python
# Correlation Heatmap of Numeric Features(cleaned data frame)
numeric_df = dfc.select_dtypes(include=['float64', 'int64'])

# Compute the correlation matrix
corr = numeric_df.corr()

# Set up the plot
plt.figure(figsize=(12, 8))

# Create a heatmap of the correlation matrix
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)

# Customize the plot
plt.title('Correlation Heatmap of Numeric Features')
plt.show()
```
